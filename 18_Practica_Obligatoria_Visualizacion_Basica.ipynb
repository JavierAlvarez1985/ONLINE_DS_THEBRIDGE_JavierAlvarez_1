### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
# 1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner
# el argumento ha a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
```

```python
ciudades = df_california.groupby("city")["population_total"].sum().sort_values(ascending=False).head(5).index.to_list()
```

```python
poblacion = df_california.groupby("city")["population_total"].sum().sort_values(ascending=False).head(5) / 1000
```

```python
poblacio = poblacion.to_list()
```

```python
colores = ["blue", "grey", "red", "olive", "brown"] 
plt.figure(figsize = (6,6))
plt.bar(ciudades, poblacion, color = colores, alpha = 0.5)
plt.title("Ciudades mas pobladas de California", fontsize=8)
plt.xlabel("Ciudades", fontsize=7)
plt.ylabel("Habitantes (miles)", fontsize=7)
plt.ylim(0, 5000)
for i, valor in enumerate (poblacion):
    plt.text(i, valor, str(valor), ha="center", fontsize=8, va="bottom")
```

```python
# 2 Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
```

```python
plt.figure(figsize=(6,6))
plt.hist(df_california["area_total_km2"], bins=50, color="steelblue", alpha = 0.7)
plt.xlabel("Superficie (km2)", font="Bradley Hand itc")
plt.ylabel("Número de ciudades", font="Bradley Hand itc")
plt.legend(title = "Hist Area", frameon=True);
```

```python
# 3 El scatter de área de agua de un ciudad con su poblacion.
```

```python
outlier1 = df_california.loc[df_california["area_water_km2"].idxmax()]
outlier2 = df_california.loc[df_california["population_total"].idxmax()]
```

```python
x1, y1 = outlier1["area_water_km2"], outlier1["population_total"] /1000
x2, y2 = outlier2["area_water_km2"], outlier2["population_total"] /1000
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df_california["area_water_km2"], df_california["population_total"] / 1000, alpha = 0.5, color = "green", s=12)
plt.annotate("Outliers", fontsize=7, xy=(x1, y1), xytext=( +60, 3500),
             arrowprops=dict(facecolor="k", width = 0.2, headwidth = 3))
plt.annotate("Outliers", fontsize=7, xy=(x2, y2), xytext=( +60, 3500),
             arrowprops=dict(facecolor="k", width = 0.2, headwidth = 3))
plt.xlabel("Area de agua en la ciudad (km2)", fontsize=7)
plt.ylabel("Poblacion (miles de habitantes)", fontsize=7);
```

```python
# 4 El boxplot de la altura de las ciudades.
```

```python
plt.figure(figsize=(6,6))
plt.boxplot(df_california["elevation_m"], vert = False, patch_artist = True, boxprops = dict(facecolor = "red"),
           medianprops = dict(color ="k"), flierprops=dict(markersize=4))
plt.title("Diagrama de caja de la altitud de las ciudades de California", fontsize=8)
plt.xlabel("Altitud (m)", fontsize=7)
```

```python
# 5 las 4 figuras juntas
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Gráfico de barras
axs[0, 0].bar(ciudades, poblacion, color = colores, alpha = 0.5)
axs[0, 0].set_title("Ciudades mas pobladas de California", fontsize=8)
axs[0, 0].set_xlabel("Ciudades", fontsize=7)
axs[0, 0].set_ylabel("Habitantes (miles)", fontsize=7)
axs[0, 0].set_ylim(0, 5000)
for i, valor in enumerate (poblacion):
    axs[0, 0].text(i, valor, str(valor), ha="center", fontsize=8, va="bottom")

# Histograma
axs[0, 1].hist(df_california["area_total_km2"], bins=50, color="steelblue", alpha = 0.7)
axs[0, 1].set_xlabel("Superficie (km2)", font="Bradley Hand itc")
axs[0, 1].set_ylabel("Número de ciudades", font="Bradley Hand itc")
axs[0, 1].legend(title = "Hist Area", frameon=True);

# Gráfico de dispersión
axs[1, 0].scatter(df_california["area_water_km2"], df_california["population_total"] / 1000, alpha = 0.5, color = "green", s=12)
axs[1, 0].annotate("Outliers", fontsize=7, xy=(x1, y1), xytext=( +60, 3500),
             arrowprops=dict(facecolor="k", width = 0.2, headwidth = 3))
axs[1, 0].annotate("Outliers", fontsize=7, xy=(x2, y2), xytext=( +60, 3500),
             arrowprops=dict(facecolor="k", width = 0.2, headwidth = 3))
axs[1, 0].set_xlabel("Area de agua en la ciudad (km2)", fontsize=7)
axs[1, 0].set_ylabel("Poblacion (miles de habitantes)", fontsize=7);

# Boxplot
axs[1, 1].boxplot(df_california["elevation_m"], vert = False, patch_artist = True, boxprops = dict(facecolor = "red"),
           medianprops = dict(color ="k"), flierprops=dict(markersize=4));
axs[1, 1].set_title("Diagrama de caja de la altitud de las ciudades de California", fontsize=8)
axs[1, 1].set_xlabel("Altitud (m)", fontsize=7)

plt.tight_layout()
plt.show()
```
